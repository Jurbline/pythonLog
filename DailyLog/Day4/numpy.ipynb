## ndadday 与 Python 原生list运算效率对比

```python
import random
import time

import numpy as np

a = []
for i in range(100000000):
    a.append(random.random())
print('随机完毕')
t1 = time.time()
sum1 = sum(a)
t2 = time.time()
b = np.array(a)
print('转换完毕')
t4 = time.time()
sum2 = np.sum(b)
t5 = time.time()
print('原生list求和耗时：', t2 - t1)
print('numpy数组求和耗时：', t5 - t4)
```

## 调整数组形状

```python
four = np.array([[1, 2, 3], [4, 5, 6]])

print(four)

four1 = four
print(id(four))
four.shape = (3, 2)
print(four)
print(id(four))
print(id(four1))
four1.shape = (2, 3)
print(four1)
print(id(four1))
print(id(four))
```

```python
#拓展随机生成小数
#使用python语法、保留两位
print(round(random.random(), 2))
#np.float16 半精度浮点数
#np.float32 单精度浮点数
#np.float64 双精度浮点数
arr = np.array([random.random() for i in range(10)])
print(arr)
print(arr.dtype)
print(arr.itemsize)
#取小数点后两位
print(np.round(arr, 2))
```

```python
#列表不可以和整数进行相加
[1, 2, 3] + 5  #TypeError: can only concatenate list (not "int") to list
```

```python
#由于numpy的广播机制，可以对不同形状的数组进行运算，加减乘除的值被广播到所有元素上进行运算
t1 = np.arange(24).reshape((6, 4))
print(t1)
print('-' * 20)
t2 = t1.tolist()
print(t1 + 2)
# print(t2 + 2)  #不能对列表进行加整数操作
print('-' * 20)
print(t1 * 2)
print('-' * 20)
print(t1 / 2)
```

```python
#形状相同，才可以进行加减乘除
t1 = np.arange(24).reshape((6, 4))
t2 = np.arange(100, 124).reshape((6, 4))
print(t1)
print(t2)
print(t1 + t2)
print(t1 - t2)
print(t1 * t2)
```

```python
#shape不同，不能进行加减乘除
t1 = np.arange(24).reshape((6, 4))
t2 = np.arange(100, 124).reshape((4, 6))
print(t1)
print(t2)
print(t1 + t2)  #ValueError: operands could not be broadcast together with shapes (6,4) (4,6)
print(t1 - t2)  #ValueError: operands could not be broadcast together with shapes (6,4) (4,6)
print(t1 * t2)  #ValueError: operands could not be broadcast together with shapes (6,4) (4,6)
```

```python
#练习轴
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print('-' * 20)
print(np.sum(a, axis=0))  # 按哪个轴求和，哪个轴就没了
print('-' * 20)
print(np.sum(a, axis=1))
print('-' * 20)
print(np.sum(a))
```

```python
#索引和切片
import numpy as np

a = np.arange(10)

#如果只放置一个参数，如[2]，将返回与该参数对应的元素
print(a[0], a)

#如哦为[2:]，表示从索引2开始的所有元素
print(a[2:])

print(a[2:8:2])  #切片是左闭右开的，即[2,8)
```

```python
#二维切片
import numpy as np

t1 = np.arange(24).reshape((4, 6))
print(t1)
print('-' * 20)
print(t1[1])  #取一行（一行代表是一行数据，索引也是从0开始的） print(t1[1, :])
print('-' * 20)
print(t1[1:])  #取连续的多行
print('-' * 20)
print(t1[1:3, :])  #取多行多列
print('-' * 20)
print(t1[[0, 2, 3]])  #取多行多列
```
