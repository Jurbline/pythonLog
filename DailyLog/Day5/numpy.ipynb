## 数据的添加，删除与去重

```python
#numpy.append() 函数在数组的末尾添加值，追加操作会分配整个数组，并把原来的数组复制到新数组中
#此外，输入数组的维度必须匹配否则将生成ValueError
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])

print('第一个数组：')
print(a)
print('\n')

print('使用append()函数添加元素：')
print(np.append(a, [7, 8, 9]))  #ndarray会被展平
print('\n')
print(a)
```

```python
b = np.array([[7, 8, 9]])
print(b.ndim)
print(b)
```

```python
print('沿轴0添加元素：')
print(np.append(a, [[7, 8, 9]], axis=0))
print('\n')
print(a)
print('沿轴1添加元素：')
print(np.append(a, [[5, 5, 5], [7, 8, 9]], axis=1))
print(a)
```

```python
#numpy.insert() 函数在给定索引之前，沿给定轴在输入数组中插入值 如果值的类型转换为要插入，则它与输入数组不同
#插入没有原地的，函数会返回一个新数组，此外，如果未提供轴，则输入数组会被展开
import numpy as np

a = np.array([[1, 2], [3, 4], [5, 6]])

print('第一个数组：')
print(a)
print('\n')

print('未传递轴参数，输入数组会被展开：')
print(np.insert(a, 1, [11, 22]))
print('-' * 50)
print(np.insert(a, 1, [11, 22], axis=0))

print('沿轴0广播：')
print(np.insert(a, 1, 11, axis=0))
print('\n')

print('沿轴1广播：')
print(np.insert(a, 1, [1, 2, 5], axis=1))
```

```python
#numpy.delete() 函数删除数组中的元素，删除操作会分配整个数组，并把原来的数组复制到新数组中，与insert()函数的情况类似
#如果未提供轴参数，则输入数组会被展开
import numpy as np

a = np.arange(12).reshape((3, 4))

print('第一个数组：')
print(a)
print('\n')

print('未传递轴参数，输入数组会被展开：')
print(np.delete(a, 5))
print('\n')
print(a)
print('删除第一行：')
print(np.delete(a, 1, axis=0))
print('\n')

print('删除第一列：')
print(np.delete(a, 1, axis=1))
```

# numpy的数学计算

```python
import numpy as np

score = np.array([[80, 88], [82, 81], [75, 81]])
print(score)
print(score.shape)
print('\n')

#获取所有数据的最大值
result = np.max(score)
print(result)
print('\n')

#获取某一个轴上的最大值，对应的轴就消失
result = np.max(score, axis=1)
print(result)

#获取最小值
result = np.min(score)
print(result)

#获取某一个轴上的最小值
result = np.min(score, axis=0)
print(result)

#数据的比较
result = np.maximum([-2, -1, 0, 1, 2], 0)  #第一个参数中的每一个数与第二个参数比较返回大的
print(result)
result = np.minimum([-2, -1, 0, 1, 2], 0)  #第一个参数中的每一个数与第二个参数比较返回小的
print(result)
result = np.maximum([-2, -1, 4, 1, 2], [1, 2, 3, 4, 5])
print(result)
#接受的两个参数，也可以大小一致；第二个参数只是一个单独的值时，其实是用到了维度的广播机制
```

```python
#求平均值
import numpy as np

result = np.mean(score)  #求所有数据的平均值
print(result)
result = np.mean(score, axis=0)  #求某一个轴上的平均值
#保留两位小数输出
print(np.round(result, 2))
#不用round()函数也可以保留两位小数
print('%.2f' % result[0])
#用f''字符串格式化输出
print(f'{result[0]:.2f}')

print('-' * 50)
#求前缀和
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
print(arr.cumsum(axis=0))  #沿轴0求前缀和
'''
[1, 2, 3]------>    |1    |2    |3    |
[4, 5, 6]------>    |5=1+4|7=2+5|9=3+6|
'''
print(arr.cumsum(axis=1))  #沿轴1求前缀和
'''
[1, 2, 3]------>    |1    |2+1    |3+2+1|
[4, 5, 6]------>    |4    |4+5    |4+5+6|
'''
```

```python
#argmin()函数返回数组中最小值的索引
import numpy as np

score = np.array([[80, 88], [82, 81], [75, 81]])
print(score)
result = np.argmin(score, axis=0)
res = np.min(score, axis=0)
print(result, res)

print('-' * 50)
#求每一列的标准差（这里是总体标准差）
#标准差是一组数据平均值分散程度的一种度量，一个较大的标准差，代表大部分数值和其平均值之间差异较大
#一个较小的标准差，代表这些数据较接近平均值反映出数据的波动稳定情况，越大表示波动越大，越不稳定
result = np.std(score, axis=0)
print(result)

#极值
result = np.ptp(score, axis=None)  #就是最大值和最小值的差值
print(result)
```

## 数组的拼接

```python
#有时候我们需要将两个数据加起来一起研究分析，我们就可以将其进行拼接然后分析
import numpy as np

#根据轴连接的数组序列，concatenate没有改变数组的维度
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = np.array([[9, 10], [11, 12]])

#要求a,b两个数组维度相同
print('沿轴0连接：')
print(np.concatenate((a, b), axis=0))
print('-' * 50)

print('沿轴1连接：')
print(np.concatenate((a, b), axis=1))
print('-' * 50)

#使用stack函数可以将数组堆叠起来，堆叠的数组的维度是原数组的维度+1
print(np.stack((a, b, c), axis=0).shape)
print(np.stack((a, b, c), axis=1).shape)
print(np.stack((a, b, c), axis=2).shape)
```

## numpy inf和nan

```python
#C 语言中表示最大的正整数值是 0x7FFF FFFF，最小的负整数值是 0x8000 0000，
#查阅资料后，发现inf表示无穷大，需要使用float('inf‘)函数来转化，那么对应的就有 float('-inf')表示无穷小了
#这样你就可以使用任意数来判断和它的关系了。
#那什么时候会出现inf呢？比如一个数字除以0，Python中会报错，但是numpy中会是一个inf或者-inf
#但是在 numpy 中，inf 表示正无穷大，负无穷大分别用 np.inf 和 -np.inf 表示。
#numpy 中还定义了两个特殊的浮点值，分别是 np.nan 和 np.inf。
#np.nan 表示 Not a Number，表示一个非数字值，例如 0/0 或者 log(-1)。
#np.inf 表示正无穷大，负无穷大分别用 np.inf 和 -np.inf 表示。
```

```python
import numpy as np

t = np.arange(24).reshape(4, 6).astype('float')
#将数组中的一部分替换为nan
print(t)
t[1, 3:] = np.nan
print(t)
print('-' * 50)
print(t.shape)
print(id(t))
```

```python
#遍历每一列，然后判断每一列是否有nan
t = np.arange(24).reshape(4, 6).astype('float')
t[1, 3:] = np.nan
for i in range(t.shape[1]):
    #获取当前列数据
    temp_col = t[:, i]

    #判断是否有nan
    # nan_num = np.count_nonzero(temp_col != temp_col)
    nan_num = np.count_nonzero(np.isnan(temp_col))
    if nan_num != 0:
        #将这一列不为nan的数据拿出来，并计算平均值
        temp_col_not_nan = temp_col[temp_col == temp_col]
        print(temp_col_not_nan)
        #将nan替换成这一列的平均值
        temp_col[np.isnan(temp_col)] = np.mean(temp_col_not_nan)

print(t)
print(id(t))
```

```python
#对换数组的维度
import numpy as np

a = np.arange(12).reshape(3, 4)
print('原数组：')
print(a)
print('\n')

print('对换数组的维度：')
print(np.transpose(a))
print(a)
print('\n')

#与transpose()函数一样
a = np.arange(12).reshape(3, 4)
print('原数组：')
print(a)
print('\n')

print('转置数组')
print(a.T)
```

```python
#函数用于交换数组的两个轴，交换后的数组的形状与原数组相同，但轴的顺序发生了变化
t1 = np.arange(24).reshape(4, 6)
re1 = t1.swapaxes(1, 0)

print('原数组：')
print(t1)
print('\n')
print(re1.shape)
print('交换后的数组：')
print(re1)
```
